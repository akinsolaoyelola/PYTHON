```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Covid_Carbon_GDP =pd.read_csv("covid_carbon_gdp_2020.csv")
Covid_Carbon_GDP
```

```python
Covid_Carbon_GDP.describe()
```

```python
Covid_Carbon_GDP.info()
```

```python
font =16
fig =plt.figure(figsize=(12,12))
axx =fig.add_axes([0, 0, 1, 1])
axx.plot(Covid_Carbon_GDP['Year'], '--r')
axx.plot(Covid_Carbon_GDP['Cumulative_cases'], '-ob')
axx.plot(Covid_Carbon_GDP['Cumulative_deaths'], '-+g')
axx.legend(['Year','Cumulative_cases','Cumulative_deaths'], fontsize=font)
axx.set_xlabel('Country', fontsize=font)
axx.set_ylabel('Cumulative_deaths', fontsize=font)
axx.set_title('Covid_Carbon_GDP.info',fontsize=font)
plt.show()
```

```python
font =16
fig =plt.figure(figsize=(12,12))
axx =fig.add_axes([0, 0, 1, 1])
axx.plot(Covid_Carbon_GDP['Country'], '--r')
axx.plot(Covid_Carbon_GDP['Yearly_Emission'], '-ob')
axx.legend(['Country','Yearly_Emission'], fontsize=font)
axx.set_xlabel('Country', fontsize=font)
axx.set_ylabel('Yearly_Emission', fontsize=font)
axx.set_title('Covid_Carbon_GDP.info',fontsize=font)
plt.show()
```
