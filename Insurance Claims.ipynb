#Exploratory Data Analysis(EDA)

```python
#Analyse Claim Frequency Per Policy Year
```

```python
filtered_claim_frequency_by_year = filtered_df_claims.groupby('YEARNAME')['CLAIMNO'].count().reset_index()
filtered_claim_frequency_by_year
```

#Comment
#The years "2018 to 2019" and "2021 to 2022" observed the peak in the number of claims, registering 249 and 245 claims respectively. There was a significant decrease in claims between '2019 to 2020" and "2020" to "2021", with 152 and 113 claims.

```python
fig, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(x='YEARNAME', y='CLAIMNO', data=filtered_claim_frequency_by_year, marker='o', ax=ax,linestyle='dotted', color='magenta')

# Adding labels, title, and a grid
ax.set_title('Claims Frequency by Policy Year')
ax.set_xlabel('Policy Year')
ax.set_ylabel('Number of Claims')
ax.grid(True)

plt.show()
```

```python
#Analyse Average Notification Delay Per Policy Year
average_delay_by_year = round(filtered_df_claims.groupby('YEARNAME')['DELAYNUM'].mean().reset_index(),0)
average_delay_by_year
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(x='YEARNAME', y='DELAYNUM', data=average_delay_by_year, marker='*', ax=ax,linestyle='solid', color='deepskyblue')\
.set(xlabel='Policy Year',ylabel='Average Notification delays',title='Average Notification Delay Per Policy Year')
```

#Comment

The year "2016 to 2017" experienced a significant increase in notification delay, averaging 29 days. However, this trend did not persist, as subsequent years saw a marked reduction in delay times, culminating in an average of just 6 days in "2022 to 2023."

```python
#Analyse Average Incurred Costs Per Policy Year
```

```python
average_incurredcosts_by_year = round(filtered_df_claims.groupby('YEARNAME')['Incurred Cost'].mean().reset_index(),0)
average_incurredcosts_by_year
```

```python
sns.barplot(x='YEARNAME', y='Incurred Cost', data=average_incurredcosts_by_year)\
.set(xlabel='Policy Year',ylabel='Average Incurred Cost',title='Average Incurred Cost Per Policy Year')
plt.xticks(rotation=45)

plt.show()
```

#Comment

The average claim costs have varied each year, suggesting that the nature and severity of incidents fluctuate annually.
Despite having fewer claims, the year 2020-2021 saw a substantial increase in the average claim cost. This suggests that the claims during this period were of higher severity, which could be due to significant accidents or incidents involving costly vehicles or assets.
